```python
## for loop
```

```python
verbs=['play','go','sing','dance','swim']
print(verbs)
```

```python
for x in verbs:
    print(x.title())
```

```python
for x in verbs:
    if x=='play':
        print(x.upper())
    else:
        print(x.title())
```

```python
for x in verbs:
    if x=='swim':
        print(x.lower())
    else:
        print(x.upper())
```

```python
for i in range(1,11):
    print('2*',i,'=',i*2)

print('')

for i in range(1,11):
    print('125',i,'=',125**i)
    
print()

for i in range(1,9):
    print('25',i,'=',i^25)
```

```python
## Introducton to Functions
```

```python
def sem():
    print("Hello User,")
    
sem()
```

```python
def sem_person(name):
    print("Hello User, " + name.lower())
    
sem_person("M.V.N Pavan Kumar")
```

```python
def sem_person(name,Company):
    print("Hello User, " + name.upper() + ", Welcome To " + Company.upper())
    
sem_person("M.V.N Pavan Kumar","Amazon")
```

```python
def wish(user):
    print(f"Hello good to have you, {user.upper()}")
    
wish("Pavan")
```

```python
def test():
    f_name="pavan"
    l_name="kumar"
    print(f"my complete name is {f_name.upper()} {l_name.upper()}")
    
test()
```

```python
## Introduction to Arguments
```

```python
1. positional arguments ---->
```

```python
def describe_pets(animal_type,pet_name):
    print(f"I have a {animal_type}")
    print(f"my {animal_type}'s name is {pet_name}")
    
describe_pets("Dog","Tom")
```

```python
describe_pets("Armadrillo","jennifer")
```

```python
2. Keyword Argument
```

```python
def describe_pets(animal_type,pet_name):
    print(f"I have a {animal_type}")
    print(f"my {animal_type}'s name is {pet_name}")
    
describe_pets(animal_type="Dog",pet_name="Tom")
```

```python
describe_pets(pet_name="Dog",animal_type="Buzz")
```

```python
describe_pets("Dog",pet_name="Buzz")
```
